```python
import networkx as nx
import matplotlib.pylab as plt 
```

```python
G = nx.Graph()

G.add_node(1)
G.add_nodes_from([2,3])
G.add_edges_from([(1,2), (2,3), (3,1)])
nx.draw(G,with_labels=True, font_weight='bold')
plt.show()
```

```python
G = nx.empty_graph(10)
G.add_edges_from([(1,2), (2,3), (3,1)])
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
G = nx.Graph()

G.add_nodes_from(["Romi", "Mai", "Sol", "Yam"])
G.add_edges_from([("Romi", "Mai"), ("Mai", "Sol"), ("Sol", "Yam"),("Yam","Romi")])

nx.draw(G,
        width=5,
        node_color="gold",
        node_size=1000,
        with_labels=True
        )

plt.show()
```
